```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
data= pd.read_csv("/content/data-27.csv")
data.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1,inplace=True)
data
```

# **Basic EDA**

```python
data["close"].describe()
```

```python
data["close"].std()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=data["datetime"], y=data["close"], name = "Closing Price", line = dict(color='blue', width=1)))
fig.update_layout(title=dict(text="CLOSING PRICE",x=0.45,y=0.9,font=dict(size=30)),yaxis_title="Closing Price",xaxis_title="Date")
fig.show()
```

# **PART 1**

```python
data["SMA"]=data["close"].rolling(window=20).mean()
data
```

```python
data["EMA_20"]=data["close"].ewm(span=20).mean()
data["EMA_50"]=data["close"].ewm(span=50).mean()
data
```

# **Part 2**

```python
data["datetime"] = pd.to_datetime(data["datetime"])  ## changed the "datetime" column to datetime format as it was showing error that it is not in datetimeindex format...
data.set_index("datetime", inplace=True)  ### "datetime" is set as index to use resample....
data=data.resample("6ME").std()
data
```

```python
print("The maximum volatility is with std.deviation:",data["close"].max())
```
